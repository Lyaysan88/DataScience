## 2. Регуляризация логрегрессии

### a. Логрегрессия по умолчанию

1. Обучите baseline-модель, только используя параметры `random_state=21`, `fit_intercept=False`.
2. Воспользуетесь K-fold кросс-валидацией со стратификацией, количество сплитов для оценки accuracy = `10`.

Результат кода, в котором вы обучили модель и оценили ее качество, должен быть точно такой же, как и ниже:

```

```

Воспользуйтесь `%%time`, чтобы понять, насколько долго выполнялась эта ячейка с кодом.

````python
def cross_val(estimator, X, y, n_splits):
  score = []
  skf = StratifiedKFold(n_splits=n_splits)
  for train_index, valid_index in skf.split(X, y):
    X_train, X_valid = X.iloc[train_index], X.iloc[valid_index]
    y_train, y_valid = y.iloc[train_index], y.iloc[valid_index]
    estimator.fit(X_train, y_train)
    score.append(accuracy_score(y_valid, estimator.predict(X_valid)))
    print( f'train -> {accuracy_score( y_train, estimator.predict( X_train )):.5f}  |  valid -> {accuracy_score( y_valid, estimator.predict( X_valid )):.5f}' )
  
  print(f'Average accuracy on crossval is {np.mean(score):.5f}')  
  print(f'Std is {np.std(score):.5f}')
````

````python
%%time
cross_val(LogisticRegression(random_state=21, fit_intercept=False), X_train, y_train, 10)
````

### b. Оптимизация параметров регуляризации

В ячейках ниже попробуйте разные значения `penalty`: `none`, `l1`, `l2`. Если потребуется, то вы также можете менять значения `solver`. Посмотрите, как меняется качество модели.

````python
solver_penalty  = { 'newton-cg': [ 'none', 'l2' ],
                    'lbfgs'    : [ 'none', 'l2' ],
                    'liblinear': [ 'l1'  , 'l2' ],
                    'sag'      : [ 'none', 'l2' ],
                    'saga'     : [ 'none', 'l1', 'l2' ] }
````

````python
for solver in solver_penalty:
    for penalty in solver_penalty[solver]:
        print( f'solver -> {solver}, penalty = {penalty}')
        cross_val( LogisticRegression( random_state = 21, fit_intercept = False, 
                                      solver = solver, penalty = penalty ), X_train, y_train, 10 )
        print()
````

Best LogisticRegression: solver = lbgfs, penalty = none
Average accuracy on crossval is 0.62537 
Std is 0.03302

### b. Оптимизация параметров регуляризации
