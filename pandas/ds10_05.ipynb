```python
surname=pd.read_json('/content/drive/MyDrive/School21/day10/surname.json', orient = 'records') #Обогатим датафрейм данными из другого источника
surname
```

```python
surname.columns = surname.iloc[0]
surname = surname.drop(index = [0], axis = 0)
surname
```

```python
surname['NAME'].where(surname['NAME'].str.replace(' ','').str.isalpha(), np.NaN, inplace = True) #Создадим новую series с фамилиями
```

```python
owners = pd.DataFrame.sample( surname['NAME'], len( fines.CarNumber.unique()), random_state = 21, replace = True )
owners
```

```python
owners = pd.DataFrame( {'CarNumber': fines.CarNumber.unique(), 'SURNAME': owners} ).reset_index(drop=True)
owners
```

```python
obs_5 = fines.copy()
obs_5 = obs_5[0:5]
obs_5['CarNumber'] = 'A'+obs_5['CarNumber']
obs_5
```

```python
fines = pd.concat([fines,obs_5])
fines
```

```python
len(owners)#ДЛЯ ПРОВЕРКИ
```

```python
len(fines)#ДЛЯ ПРОВЕРКИ
```

```python
owners.drop( owners.index[len(owners)-20:len(owners)], inplace = True )
obs_3 = owners.copy()
obs_3 = obs_3[0:3]
obs_3['CarNumber'] =  'D'+ obs_3['CarNumber']
owners = pd.concat([owners,obs_3])
owners
```

```python
fines.merge( owners, how = 'inner', left_on = 'CarNumber', right_on = 'CarNumber' ) #обьединяем ДФ 4-мя способами
```

```python
fines.merge( owners, how = 'outer', left_on = 'CarNumber', right_on = 'CarNumber' )
```

```python
fines.merge( owners, how = 'left',  left_on = 'CarNumber', right_on = 'CarNumber' )
```

```python
fines.merge( owners, how = 'right', left_on = 'CarNumber', right_on = 'CarNumber' )
```

```python
pd.pivot_table( fines, values = 'Fines', index = [ 'Make', 'Model' ], columns = ['Year'], aggfunc = np.sum ) #Создадим сводную таблицу из датафрейма fines. Значения являются суммами штрафов, но с указанием всех годов
```

```python
fines.to_csv('/content/drive/MyDrive/School21/day10/fines.csv', index=False) #Сохраним датафреймы fines и owners в CSV-файлы без индекса
```
