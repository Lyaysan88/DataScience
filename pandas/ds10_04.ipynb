```python
df = pd.read_json(
    '/content/drive/MyDrive/School21/day10/auto.json',
    orient='records'
    )
df
```

```python
df.set_index(keys='CarNumber', drop=True, inplace=True); df #Carnumber в качестве индексного столбца
```

```python
df.query("Model in ['Focus', 'Corolla']")
```

```python
df.loc[['Y7689C197RUS', '92928M178RUS', '7788KT197RUS', 'H115YO163RUS', 'X758HY197RUS']]
```

```python
df.groupby("Make").agg({'Fines': 'median'})
```

```python
df.groupby(["Make", "Model"]).agg({'Fines': 'median'})
```

```python
df.groupby(["Make", "Model"]).agg({'Fines': 'count'})
```

```python
df.groupby(["Make", "Model"]).agg({'Fines': 'min'})#минимальный штраф
```

```python
df.groupby(["Make", "Model"]).agg({'Fines': 'max'})#максимальный штраф
```

```python
df.groupby(["Make", "Model"]).agg({'Fines': 'std'})#стандартное отклонение штрафов
```

```python
df.groupby(['Make', 'Model'])['Fines'].agg(['min','max','std'])
```

```python
df.groupby('CarNumber')['Fines'].count().sort_values( ascending=False) #номера автомобилей, сгруппированные по количеству штрафов в порядке убывания, мы хотим найти тех, кто чаще всего нарушал правила дорожного движения
```
