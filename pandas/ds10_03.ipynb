```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/drive/MyDrive/School21/day10/auto.csv', index_col = 'ID')
df
```

```python
df.count()#количество наблюдений
```

```python
df.drop_duplicates(subset=['CarNumber', 'Make_n_model', 'Fines'],
                   keep='last',
                   inplace=True)#удаляем дубликаты
df
```

```python
df.count()#количество наблюдений
```

```python
df.isna().sum()#отсутствующие значения
```

```python
df.dropna(axis=1,
          thresh=500,
          inplace=True)#удаляем столбцы с пропущенными значениями
df
```

```python
df.Refund = df.Refund.fillna(method='ffill', axis=0)
```

```python
df.count()#количество наблюдений
```

```python
mean_fine = df.Fines.mean()#среднее значение
```

```python
df.fillna(value={'Fines': mean_fine}, axis=0, inplace=True); df
```

```python
df.isna().sum()#отсутствующие значения
```

```python
df['count'] = df.apply(lambda x: len(x['Make_n_model'].split()),
                       axis=1)
df.query('count != 2')
```

```python
df.drop(columns='count', inplace=True)
```

Функция разделения на марку и модель

```python
def split_str(x):
  x = x['Make_n_model'].split()
  if len(x) >= 2:
    return x[0:2]
  else:
    return [x[0],pd.NA]
```

```python
df[['Make', 'Model']] = df.apply(split_str,
         axis=1,
        result_type='expand')
df
```

```python
df.isna().sum()#отсутствующие значения
```

```python
df.drop(columns='Make_n_model', inplace=True); df
```

```python
df.to_json('/content/drive/MyDrive/School21/day10/auto.json', orient='records')#сохраняем датафрейм в формате json
```
