Используя метод read_csv:

```python
import pandas as pd
import numpy as np
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
df = pd.read_csv(
    '/content/drive/MyDrive/School21/day10/feed-views.log',
    sep='\t',
    engine='python', 
    header=None, 
    skiprows=[2, 3], 
    skipfooter=2,  
    names = ['datetime', 'user'], 
    index_col = 'datetime'
    )
df
```

```python
df.index.names
```

```python
df.index.names = ['date_time'];df
```

```python
df.to_csv('/content/drive/MyDrive/School21/day10/feed-views-semicolon.log', sep=';')
```
