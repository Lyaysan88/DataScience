```python
def time_format(time):
  if time<=60:
    return f"{time} секунд"
  else:
    time/=60

  if time<=60:
    return f"{time} минут"
  else:
    time/=60 

  if time<=24:
    return f"{time} часов"
  else:
    return f"{time/24} дней"
```

```python
def get_time (fun, params):
  n_lower, time_lower, n_higher = params
  return time_format(time_lower*fun(n_higher)/fun(n_lower))
```

```python
def search(data, value):
    n = len(data)
    left = 0
    right = n - 1
    while left <= right:
        middle = (left + right) // 2
        if value < data[middle]:
            right = middle - 1
        elif value > data[middle]:
            left = middle + 1
        else:
            return middle
    raise ValueError('Value is not in the list')
    
if __name__ == '__main__':
    data = [1, 5, 3, 7, 2, 4, 9, 6, 5]
    print(search(data, 4))
```

```python
def merge(left, right):
    n = len(left)
    m = len(right)
    i = 0
    j = 0
    result = [None] * (n + m) 
    while i + j < n + m:
        if j == m or (i < n and left[i] < right[j]):
            result[i + j] = left[i]
            i += 1
        else:
            result[i + j] = right[j]
            j += 1
            
    return result
 
def sorter(array):
 
    n = len(array)
    if n <= 1:
        return array
 
    middle = int(n // 2)
    left = sorter(array[:middle])
    right = sorter(array[middle:])
 
    return merge(left, right)
```

```python
def sorter(data):
    swapped = True
    while swapped:
        swapped = False
        for i in range(len(data)-1):
            if data[i] > data[i+1]:
                data[i], data[i+1] = data[i+1], data[i]
                swapped = True
    
if __name__ == '__main__':
    data = [9, 1, 7, 6, 2, 8, 5, 3, 4, 0]
    sorter(data)
    print(data)
```

```python
def permutation(data, n):
    if n == 1:
        print(data)
        return
    
    for i in range(n):
        permutation(data, n - 1)
        if n % 2 == 0:
            data[i], data[n-1] = data[n-1], data[i]
        else:
            data[0], data[n-1] = data[n-1], data[0]
    
if __name__ == '__main__':
    data = [1, 2, 3]
    permutation(data, len(data))
```
